# Uniforme
* todas tienen la misma probabilidad de ocurrencia. ver la frmula del lado izquierdo: 1/b-a
* en la funcin hay que especificar punto de origen y punto final, como parmetros



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a, b, n = 1, 100, 10000
data = np.random.uniform(a,b,n)
plt.hist(data)
```
