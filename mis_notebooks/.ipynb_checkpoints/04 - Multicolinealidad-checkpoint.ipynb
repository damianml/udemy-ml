# Multicolinealidad

es cuando dos variables predictoras (las equis) tiene correlación entre sí.

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../datasets/ads/Advertising.csv')
```

### se observa una correlación significativa, 0.35, entre Radio y Newspaper (dos features predictoras), lo que prejudica al modelo de regresión

```python
sns.heatmap(data.corr(), annot=True)
```

## hay que calcular el VIF: factor de inflación de la varianza = 1-R2
### posibles resultados
* **VIF = 1**: las variable no están correlacionadas.
* **VIF <= 0.5**: correlación moderada, aunque pueden seguir siendo parte del modelo
* **VIF > 0.5**: debe descartarse la correlación eliminando alguna de las variables
