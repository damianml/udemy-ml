**cargar un fichero con open**

```python
df3 = open(main_path + "/customer-churn-model/Customer Churn Model.txt")
```

```python
#con strip quito los caracteres en blanco al principio y final de la linea de texto
cols = df3.readline().strip().split(",")
n_cols = len(cols)
```

```python
main_dict = {} #creo un diccionario vac√≠o
#recorro cada columna de "cols", creando un diccionario en donde la key es el nombre
#de la columna y como valor una lista vac√≠a:
for col in cols:
    main_dict[col] = []
main_dict
```

```python
#ahora recorro el archivo:
counter = 0
for line in df3:
    values = line.strip().split(",") #almaceno todos los valores de la fila
    #recorro las columnas:
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i]) #y agrego en cada "celda" su respetivo valor
    counter += 1
print("el dataset tiene %d fila y %d columnas"%(counter, n_cols))
```

```python
df4 = pd.DataFrame(main_dict)
df4.head()
```

```python
df3.close()
```

* **al abrir un archivo en modo "w" √©ste pierde todo su contenido**
* **utilizar el "a" (append) para abrirlo si se quiere agregar informaci√≥n al final**



**otro trabajo de lectura/escritura**

```python
#creo el acceso al archivo original y al de salida:
infile = main_path + "/customer-churn-model/Customer Churn Model.txt"
outfile = main_path + "/customer-churn-model/my tab Customer Churn Model.txt"
```

```python
counter = 0
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
            counter += 1
print("se copiaron %d filas."%(counter))
```
