```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("../datasets/customer-churn-model/Customer Churn Model.txt")
data.describe(include='all')
```

```python
plt.figure(figsize=(15,9))
sb.heatmap(data.corr(), annot=True)
```

```python
%matplotlib inline
```

## scatterplot

```python
data.plot(kind='scatter', x='Day Mins', y='Day Charge', color='red')
```

```python
data.plot(kind='scatter', x='Night Mins', y='Night Charge', color='black')
```

```python
figure, axs = plt.subplots(2,2,sharey=True, sharex=True)
data.plot(kind='scatter', x='Day Mins', y='Day Charge', color='red', ax=axs[0][0])
data.plot(kind='scatter', x='Night Mins', y='Night Charge', color='black', ax=axs[0][1])
data.plot(kind='scatter', x='Day Calls', y='Day Charge', color='orange', ax=axs[1][0])
data.plot(kind='scatter', x='Night Calls', y='Night Charge', color='olive', ax=axs[1][1])
```

## histograma de frecuencias

```python
bins = int(np.ceil(1+np.log2(3333))) #regla de sturges
plt.hist(data['Day Calls'], bins=bins, color='skyblue') #el bins podría ser: bins=[0,30,60,90,120,150,180] (manual)
plt.xlabel('nro de llamadas')
plt.ylabel('frecuencia')
plt.title('nro de llamadas', fontsize=15)
```

## boxplot

```python
plt.boxplot(data['Day Calls'])
plt.ylabel('nro de llamadas')
```

```python
ric = data['Day Calls'].quantile(0.75) - data['Day Calls'].quantile(0.25)
ric
```

```python
inferior = data['Day Calls'].quantile(0.25) - 1.5*ric
inferior
```

```python
superior = data['Day Calls'].quantile(0.75) + 1.5*ric
superior
```
