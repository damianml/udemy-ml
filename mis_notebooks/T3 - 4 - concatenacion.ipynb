## tipos de joins

* <font color='red'>inner join</font>: devuelve los registros que están en ambos dataframes
    * dataA = 60 registros
    * dataB = 50 registros
    * si ambos comparten 30 filas, el dataset resultante tendrá esas 30 filas
    
* <font color='red'>left join</font>: devuelve los registros que están en el df de la izquierda
    * dataA = 60 registros
    * dataB = 50 registros
    * si ambos comparten 30 filas, el dataset resultante tendrá esas 60 filas
    * el resultado de la columna a la derecha contendrá NA para los valores que no tienen correspodencia con la columna izquierda
* <font color='red'>left join</font>: idem al anterior, pero a la inversa
* <font color='red'>outer join</font>: devuelve las filas de ambos, colocando NA donde no hay correspondencia entre columnas
    * dataA = 60 registros
    * dataB = 50 registros
    * si ambos comparten 30 filas, devolverá: 60 + 50 - 30 = 80 filas

**ejemplos**

```python
#selecciono a 6 atletas al azar:
out_athletes = np.random.choice(data_main['Athlete'], size=6, replace=False)
print("atletas eliminados ", out_athletes)
```

```python
#creo un dataframe con los atletas que NO están en out_athletes y también quito al que tiene más medallas
data_country_dlt = data_country_sin_dp[(~data_country_sin_dp['Athlete'].isin(out_athletes)) 
                                    & (data_country_sin_dp['Athlete'] != "Michael Phelps")]
data_sports_dlt = data_sports[(~data_sports['Athlete'].isin(out_athletes)) 
                                    & (data_sports['Athlete'] != "Michael Phelps")]
data_main_dlt = data_main[(~data_main['Athlete'].isin(out_athletes)) 
                                    & (data_main['Athlete'] != "Michael Phelps")]
```

```python
#comparo la longitudes, para verificar que los 7 atletas no estén en el nuevo dataframe
len(data_country_sin_dp) - len(data_country_dlt)
```

## vamos con los ejemplos de cada join

```python
#data_main contiene toda la info; en data_country_dlt faltan 7 atletas
merged_inner = pd.merge(left=data_main, right=data_country_dlt, how='inner',
                        left_on='Athlete', right_on='Athlete')
merged_inner.head(4)
```

```python
#aprecen los atletas que NO tienen países que ya fueron eliminados en los procedimientos anteriores
merged_left = pd.merge(left=data_main, right=data_country_dlt, how='left',
                        left_on='Athlete', right_on='Athlete')
merged_left.head(4)
```

```python
#están todos los atletas del data_main, y los países faltantes de completaron con NaN
print(len(merged_left))
print(len(data_main))
```

```python
#cantidad de nuls en el nuevo dataframe mergeado
merged_left['Country'].isnull().sum()
```
