# <font color='green'>Random Forest</font>

* oob_score: al muestreo aleatorio lo realiza el algoritmo (en True)
* n_estimators: número de árboles del modelo
* nota: con n_estimators en 10, obtuve un promedio de 21; con ese valor en 30, bajó al 11%

```python
forest = RandomForestRegressor(n_jobs=2, oob_score=True, n_estimators=500)
forest.fit(x, y)
```

**analizo los resultados**

```python
data['rforest_pred'] = forest.oob_prediction_
data[['rforest_pred', 'medv']]
```

**calculo el error cuadrático medio**

```python
data['rforest_error'] = (data.rforest_pred - data.medv)**2
data.rforest_error.sum()/len(data)
```

**importancia otorgada a cada feature** (sigue la misma lógica que el árbol de decisión)

```python
list(zip(predictors, forest.feature_importances_))
```
