```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor, export_graphviz
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import StratifiedKFold
import os
from graphviz import Source
```

```python
data = pd.read_csv('../datasets/boston/Boston.csv')
data.head()
```

```python
data.describe(include='all')
```

```python
#sns.pairplot(data, height=4,kind='scatter')
```

* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per 10,000
* PTRATIO - pupil-teacher ratio by town
* B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
* LSTAT - '%' lower status of the population
* MEDV - Median value of owner-occupied homes in $1000's

```python
plt.figure(figsize=(15,12))
sns.heatmap(data.corr(), annot=True)
```

**preparo la data**

```python
colnames = data.columns.values.tolist()
predictors = colnames[0:13]
target = colnames[13]
x = data[predictors]
y = data[target]
```

**creo, fiteo y predigo**
* juego un poco con el max_depth

* criterion{mse, friedman_mse, mae}, default=mse :The function to measure the quality of a split
* splitter{best, random}, default=best
* max_depth int, default=None
* min_samples_split int or float, default=2 The minimum number of samples required to split an internal node:
* min_samples_leaf int or float, default=1 The minimum number of samples required to be at a leaf node
* min_impurity_decreasefloat, default=0.0 A node will be split if this split induces a decrease of the impurity greater than or equal to this value.

```python
regtree = DecisionTreeRegressor(criterion='mse', min_samples_split=30, min_samples_leaf=10, random_state=0, max_depth=7)
regtree.fit(x, y)
```

```python
preds = regtree.predict(x[predictors])
data['preds'] = preds
```

**primera visualizacin de las predicciones**
* ver que las preds se repiten, xq el promedio cae en ese nodo

```python
data[['preds', 'medv']]
```

```python
print("media de las predicciones ", data['preds'].mean())
print("media de los val. reales  ", data['medv'].mean())
```

```python
with open("rtree_boston.dot", "w") as dotfile:
    dot_data=export_graphviz(regtree, out_file=dotfile, feature_names=predictors, rounded=True, 
                    filled=True, special_characters=True)
    dotfile.close()
    
file = open("rtree_boston.dot")
text = file.read()
Source(text)
```

```python
cv = KFold(n_splits=10, shuffle=True, random_state=99)
for i in range(1,11):
    regtree = DecisionTreeRegressor(criterion='mse', min_samples_split=30, min_samples_leaf=10, random_state=0, max_depth=i)
    regtree.fit(x, y)
    cv.split(x)
    single_score = cross_val_score(regtree, x, y, n_jobs=1, cv=cv)
    print("max_depth {}, score {}".format(i, np.mean(single_score)*100))
    #print("    feature de importancia ", regtree.feature_importances_)
    
```
