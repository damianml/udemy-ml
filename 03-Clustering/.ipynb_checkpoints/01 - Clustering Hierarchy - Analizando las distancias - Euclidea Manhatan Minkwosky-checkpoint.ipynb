## ejemplo:

sistema de recomendacin para pelculas

**Distancia de MANHATAN (cuenta las "cuadras" de distancia, no va en lnea recta)**
* toma las distancias entre las calificaciones que cada usuario hizo de las pelculas
* las resta y obtiene el valor absoluto
* luego suma todos esos valores absolutos
* el resultado es: 9.9



**Distancia EUCLDEA**
* va en "lnea recta"
* es la raz de las (sumas de las diferencias)^2
* es la distancia ms utilizada
* notar que el valor obtenido es ms bajo que el de MANHATAN


**Distancia MINKOWSKY**
* funciona como las anteriores
* el tema es el valor de p:
    * si p = 1 = MANHATAN
    * si p = 2 = EUCLDEA
    * si el valor de p aumenta, penaliza menos las distancias. sube p, se penaliza menos x las distancias existentes
