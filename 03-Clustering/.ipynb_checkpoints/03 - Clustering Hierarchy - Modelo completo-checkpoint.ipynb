```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
np.random.seed(4711)
```

```python
#genero una matriz con dos columnas
#parmámtros: 10,0 = valor del eje equis en torno a 10, con una std de 3 horizontal y 4 vertical. 100 puntos ploteados
a = np.random.multivariate_normal([10,0], [[3,1], [1,4]], size=[100,])
b = np.random.multivariate_normal([0,20], [[3,1], [1,4]], size=[50,])
x = np.concatenate((a,b))
plt.scatter(x[:,0], x[:,1])
plt.show()
```

**librerías para medir el coeficiente de correlación de cophenet**
    * compara la distancias, dos a dos, de las muestras, con las que llevó a cabo el clustering
    * cuanto más cercano a 1, mejor será el clustering

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
z = linkage(x, method='ward') #realizo el cluster
#paso el cluster y dentro de pdis a x
c, coph_dist = cophenet(z, pdist(x))
```

```python
print("{} % de igualdad entre las distancias originales y los clusteres que se han generado".format(np.around(c,4)*100))
```

### veamos un ejemplo de cómo se forma un clúster básico

* la primera y segunda columnas son, por ejemplo, los puntos a y b
* tercera columna: distancia entre a y b
* cuarta: cantidad de puntos en ese clúster

```python
print(z[0])
print(z[50])
print(z[100])
```

```python
plt.figure(figsize=(20,10))
plt.title("Dendrograma")
plt.xlabel("Muestras")
plt.ylabel("Distancias")
#ojo con el valor 180 del color_threshold, ver bien el por qué de ese número
dendrogram(z, leaf_rotation=90., leaf_font_size=8., color_threshold=0.1*180)
plt.show()
```

**analizo los últimos 4 registros del linkage**
en la última columna se aprecian los "saltos" que va dando la clusterización. Pasa, por ejemplo, del 71 al 50, para luego ir hasta el 100 y de ahí al 150

```python
z[~4:,]
```

## cómo truncar el dendrograma

ver los parámetros en la segunda línea del dendrogram: la `p` es la cantidad de nodos; y el show_contracted para mostrar a los nodos contraídos

con "plt.axhline(y=20, c='r')" trazo una línea horizontal para establecer un corte

```python
plt.figure(figsize=(20,10))
plt.title("Dendrograma TRUNCADO")
plt.xlabel("Muestras")
plt.ylabel("Distancias")
plt.axhline(y=20, c='r')
#ojo con el valor 180 del color_threshold, ver bien el por qué de ese número
dendrogram(z, leaf_rotation=90., leaf_font_size=8., color_threshold=0.1*180,
          truncate_mode="lastp", p=15, show_leaf_counts=True, show_contracted=True)
plt.show()
```

## corte automático del cluster
### Método del Codo

```python
last = z[-10:, 2] #tomo los últimos registros del linkage, de la 3ra columna
last_rev = last[::-1] #revierto su posición
idx = range(1, len(last)+1) #armo el arange para el eje x
plt.plot(idx, last_rev, c='orange')
```

## Recuperar los clusters y sus elementos

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_d = 20 # a este valor lo saco del dendrograma, es la línea de "corte" que divide los clusters
#paso el linkage (z), la altura del corte y el criterio
clusters = fcluster(z, max_d, criterion='distance')
clusters
```

**tambien es posible indicar el nro de clústers**

```python
k=3
clusters = fcluster(z, k, criterion="maxclust")
clusters
```

**otra forma de graficar los clusters**

```python
plt.figure(figsize=(10,7))
plt.scatter(x[:, 0], x[:, 1], c=clusters, cmap='prism')
plt.show
```
