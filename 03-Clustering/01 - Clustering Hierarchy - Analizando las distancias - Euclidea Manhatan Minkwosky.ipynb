## ejemplo:

sistema de recomendación para películas

**Distancia de MANHATAN (cuenta las "cuadras" de distancia, no va en línea recta)**
* toma las distancias entre las calificaciones que cada usuario hizo de las películas
* las resta y obtiene el valor absoluto
* luego suma todos esos valores absolutos
* el resultado es: 9.9



**Distancia EUCLÍDEA**
* va en "línea recta"
* es la raíz de las (sumas de las diferencias)^2
* es la distancia más utilizada
* notar que el valor obtenido es más bajo que el de MANHATAN


**Distancia MINKOWSKY**
* funciona como las anteriores
* el tema es el valor de p:
    * si p = 1 = MANHATAN
    * si p = 2 = EUCLÍDEA
    * si el valor de p aumenta, penaliza menos las distancias. sube p, se penaliza menos x las distancias existentes
