```python
from scipy.spatial import distance_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('../datasets/movies/movies.csv', sep=";")
data.head()
```

```python
data.describe(include='all')
```

```python
#solo tomo las columnas desde la 1 en adelante, así omito cargar el user_id
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1) #manhatan
dd2 = distance_matrix(data[movies], data[movies], p=2) #euclidea
dd5 = distance_matrix(data[movies], data[movies], p=5) #minkowski
```

**creo una funcion para pasar esas matrices a un dataframe**

```python
def dd_to_df(dd, col_name):
    #coloco como index y nombre de columna lo mismo, ya que será una matriz que cruza datos
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

**voy creando las <> matrices de distancia**

```python
dd_to_df(dd1, data['user_id'])
```

```python
dd_to_df(dd2, data['user_id'])
```

```python
dd_to_df(dd3, data['user_id'])
```

# Enlaces


```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
z = linkage(data[movies], method='ward')
plt.figure(figsize=(20,10))
plt.title("Dendograma de Películas")
plt.xlabel("Usuarios de Netflix")
plt.ylabel("Distancias")
dendrogram(z, leaf_rotation=90., leaf_font_size=15)
plt.show()
```

```python
z = linkage(data[movies], method='average')
plt.figure(figsize=(20,10))
plt.title("Dendograma de Películas")
plt.xlabel("Usuarios de Netflix")
plt.ylabel("Distancias")
dendrogram(z, leaf_rotation=90., leaf_font_size=15)
plt.show()
```
