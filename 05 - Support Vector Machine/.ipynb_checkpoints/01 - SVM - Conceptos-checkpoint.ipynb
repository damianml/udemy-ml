* la más sencilla de las clasificaciones lineales de la SVM se llama "maximizador de los márgenes", o bien llamado el "perceptrón de estabilidad óptima". Esto EN CASO DE EXISTIR, ya que no siempre es así.


* de forma teórica, ver el "canal" que se genera en el hiperplano "amarillo" (d) el cual es igual a la derecha e izquierda de dicho hiperplano
* esta distancia (d) se mide desde el hiperplano hacia los puntos más cercanos a él (en el dibujo, el rojo de la izquierda y los dos azules de la derecha
* es "canal" es la frontera de decisión

Las SVM son buenas para:
* reconocimiento óptico de caracteres
* detección de caras en el enfoque de cámaras fotográficas
* reconocimiento de imágenes satelitales (separando qué zonas tienen nubes, agua, tierra, hielo, etc)
* filtros de spam para correos electrónicos

con el paso de los años, las redes neuronales profundas superaron a las SVM, ya que poseen mayor capacidad de generalización y aprendizaje.
