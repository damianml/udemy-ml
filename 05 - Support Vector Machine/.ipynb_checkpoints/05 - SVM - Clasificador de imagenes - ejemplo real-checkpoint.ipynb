# Reconocimiento facial

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.svm import SVC
from sklearn.decomposition import PCA as rPCA
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
print("imágenes de: ", faces.target_names)
print("*"*50)
print("cantidad de imágenes, píxeles, píxeles:")
print("tamaño: ", faces.images.shape)
```

```python
fig, ax = plt.subplots(5, 5, figsize=(16, 9))
for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(faces.images[i], cmap='bone')
    ax_i.set(xticks=[], yticks=[], xlabel=faces.target_names[faces.target[i]])
plt.show()
```

## para no sobrecargar al algoritmo, en SVM no se pasan todos los píxeles, sino aquellos que marcan las principales características de la imagen

## defino:
* el pca (ver qué es whiten)
* el svc (con kernel radial y el peso de las clases en "balanceado")
* y en model creo el pipeline, pasando los dos algoritmos como parámetros

```python
pca = rPCA(n_components=150, whiten=True, random_state=42)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
```

## conjunto de entrenamiento y testeo
## luego armo el diccionario de parámetros

```python
x_train, x_test, y_train, y_test = train_test_split(faces.data, faces.target, random_state=42)
```

```python
param_grid = {
    "svc__C":[0.1, 1, 5, 10, 15],
    "svc__gamma":[0.0001, 0.0005, 0.001, 0.005, 0.01], #radio del svc (son valores chicos)
}
```

## paso el model (pca y svc) al gridSearchCV
### tomo el tiempo que tarda:

```python
grid = GridSearchCV(model, param_grid)
%time grid.fit(x_train, y_train)
```

## obtengo los mejores parámetros:

```python
print(grid.best_params_)
```

## reconfiguro el algoritmo con esos valores y realizo la predicción (con x_test)

```python
classifier = grid.best_estimator_
y_pred = classifier.predict(x_test)
y_pred
```

```python
fig, ax = plt.subplots(8, 6, figsize=(16, 9))
for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(x_test[i].reshape(62, 47), cmap='bone') #en el reshape va la altura y ancho de la imagen
    ax_i.set(xticks=[], yticks=[])
    ax_i.set_ylabel(faces.target_names[y_pred[i]].split()[-1], 
                    #coloco un if... para que ponga en rojo los que NO coincidieron
                    color='black' if y_pred[i]==y_test[i] else 'red')
fig.suptitle("Predicción de Imágenes [incorrecta en rojo]", size=15)
plt.show()
```

## realizo un reporte:

* precision: especifidad, cuántas veces acertó.
* recall: sensibilidad, 1 - falsos negativos, cuántas veces que NO era la imagen la clasificó como NO
* f1: coeficiente entre ambos
* support: cantidad de ejemplos

```python
print(classification_report(y_test, y_pred, target_names=faces.target_names))
```

```python
plt.figure(figsize=(10,7))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, xticklabels=faces.target_names, yticklabels=faces.target_names, fmt='d')
```
