```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns; sns.set()
from sklearn import svm
from scipy import stats
from sklearn.datasets.samples_generator import make_blobs
```

```python
x, y = make_blobs(n_samples=50, centers=2, cluster_std=0.6, random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap='autumn')
plt.show()
```

**agregamos un punto al gráfico**

**¿dónde entraría el punto azul?**

**observar que hay muchos hiperplanos posibles, ahora hay que maximizar las distancias**

```python
xx = np.linspace(-1, 3.5)
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap='autumn')
plt.plot([0.5], [2.1], "x", color='blue', markeredgewidth=2, markersize=8)

#ahora genero las posibles líneas de división
#la "a" es la pendiente, la b es la ordenada al origen
#creo tres configuraciones de recta:
for a, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    yy = a * xx + b
    plt.plot(xx, yy, "k-")
plt.xlim(-1, 3.5)
plt.title("Varias opciones")
plt.show()
```

## maximizar el margen de clasificación

en lugar de pintar un recta, que es una línea sin dimensión, por qué no dibujar un corredor, con la anchura suficiente hasta encontrar el primer punto?

```python
xx = np.linspace(-1, 3.5)
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap='autumn')
plt.plot([0.5], [2.1], "x", color='blue', markeredgewidth=2, markersize=8)

#ahora genero las posibles líneas de división
#la "a" es la pendiente, la b es la ordenada al origen
#creo tres configuraciones de recta (los últimos (0.33,0.55,0.2)) es el valor de la "d" que se utiliza
#dentro de fill_between
for a, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yy = a * xx + b
    plt.plot(xx, yy, "k-")
    plt.fill_between(xx, yy-d, yy+d, edgecolor='none', color="#BBBBBB", alpha=0.5)
plt.xlim(-1, 3.5)
plt.title("El tunel central es el indicado")
plt.show()
```
